# Task 4

```python
user_activity =[{'user_id': '1001', 'event_type': 'login', 'duration_seconds': 20},
               {'user_id': '1002', 'event_type': 'browse', 'duration_seconds': 180},
               {'user_id': '1003', 'event_type': 'logout', 'duration_seconds': 'a2r'},
               {'user_id': '1004', 'event_type': 'purchase', 'duration_seconds': 117},
               {'user_id': '1005', 'event_type': 'login', 'duration_seconds': 35},
               {'user_id': '1006', 'event_type': 'browse', 'duration_seconds': 78},
               {'user_id': '1007', 'event_type': 'logout', 'duration_seconds': 12},
               {'user_id': '1008', 'event_type': 'browse', 'duration_seconds': -123},
               {'user_id': '1009', 'event_type': 'purchase', 'duration_seconds': 120},
               {'user_id': '1010', 'event_type': 'login', 'duration_seconds': 27},
               {'user_id': '1011', 'event_type': 'browse', 'duration_seconds': 68},
               {'user_id': '1012', 'event_type': 'login' , 'duration_seconds': 45}
               ]
```

```python
cleaned_events = []
for event in user_activity:
    duration = event.get("duration_seconds")
    if isinstance(duration, (int, float)) and duration >= 0:
        copy_event = event.copy()
        copy_event['duration_minutes'] = duration/60
        cleaned_events.append(copy_event)
print(len(cleaned_events))
```

```python
cleaned_events
```

----------------------------------------

# Task 5

```python
event_types = [event["event_type"] for event in cleaned_events]
```

```python
event_types
```

```python
event_count = {}
for etype in set(event_types):
    event_count[etype] = event_types.count(etype)
```

```python
event_count
```

```python
duration_by_type = {}

for event in cleaned_events:
    etype = event["event_type"]
    duration = event["duration_minutes"]
    if etype not in duration_by_type:
        duration_by_type[etype] = []
    duration_by_type[etype].append(duration)
average_duration = {etype: (sum(durations) / len(durations)) for etype, durations in duration_by_type.items()}

print(average_duration)
```

```python
unique_users = {event["user_id"] for event in cleaned_events}
print(len(unique_users))
```

```python
sum(event_count.values())==len(cleaned_events)
```

```python
columns = ["metric,key,value"]
```

```python
event_count
```

```python
for key, count in event_count.items():
    columns.append(f'count,{key},{count}')
```

```python
columns
```

```python
average_duration
```

```python
for key, average in average_duration.items():
    columns.append(f'average,{key},{average:.2f}')
```

```python
columns
```

```python
columns.append(f"unique_users,all,{len(unique_users)}")
```

```python
columns
```

```python
csv_string = "\n".join(columns)
print(csv_string)
```

```python
with open("user_activity.csv", "w") as csv_file:
    csv_file.write(csv_string)
```
